## Описание проекта

В связи с увеличением оттока клиентов банка, который угрожает банку разорением, его отдел маркетинга (далее заказчик) просит провести исследование данных с целью выявления сегментов отточных клиентов. Сегменты должны быть однородными, высокооточными и покрывать максимальное количество проблемных зон банка. По выделенным сегментам требуется дать рекомендации, как снизить отток клиентов из банка.

### Распределение данных в разрезе оттока клиентов

Для построения графиков распределения создам функцию.

```python
# функция для построения распределения данных в разрезе оттока клиентов
def distribution(df, hue, bins, xlim=None, ylim=None): # (датасет, столбец на основании которого разрезаю данные, корзины, ограничение по оси х, ограничение по оси у)
    for columns in df.columns:
        if columns == hue: # условие, чтобы не строить график по этому столбцу
            continue
        else:
            sns.displot(df, x=columns, hue=hue, \
                        stat='density', common_norm=False, height=7, bins=bins)
            plt.title('Распределение данных в разрезе оттока клиентов по {}'.format(columns), pad=30)
            plt.xlim(xlim)
            plt.ylim(ylim)                           
            plt.show()
```

## Проверка статистических гипотез

**В ТЗ сформулирована гипотеза № 1:**

Нулевая гипотеза: между клиентами, которые ушли и теми, которые остались нет значимой разницы в уровне дохода (столбец 'est_salary').

Альтернативная гипотеза: между клиентами, которые ушли и теми, которые остались есть значимая разница в уровне дохода (столбец 'est_salary').

**Для внесения ясности в исследование формулирую гипотезу № 2:**

Нулевая гипотеза: между клиентами, которые ушли и теми, которые остались нет значимой разницы в количестве используемых продуктов (столбец 'products').

Альтернативная гипотеза: между клиентами, которые ушли и теми, которые остались есть значимая разница в количестве используемых продуктов (столбец 'products').

Задаю пороговое значение alpha 0.05.

В задаче две генеральные совокупности по столбцу 'churn': ушел клиент или остался. Поэтому для проведения теста использую метод .ttest_ind. Предполагаю, что дисперсии у совокупностей разные, поэтому equal_var будет принимать значение False. Также поскольку в обоих столбцах, по которым буду проверять гипотезы есть выбросы, то вместо среднего арифметического буду использовать медиану.

```python
# функция для проверки гипотезы
def test_hyp(slice_1, slice_2, columns, alpha):
    # расчет медианных значений зарплаты отточных и оставшихся пользователей
    selection_outflow = slice_1
    selection_not_out = slice_2
    print('Медианное значение выборки отточных клиентов для столбца {}:'.format(columns), selection_outflow.median())
    print('Медианное значение выборки оставшихся клиентов для столбца {}:'.format(columns), selection_not_out.median())
    
    # проверка гипотезы
    alpha = alpha
    results = st.ttest_ind(selection_outflow, selection_not_out, equal_var=False)
    print('p-значение:', results.pvalue)

    if (results.pvalue < alpha):
        print("Отвергаем нулевую гипотезу")
    else:
        print("Не получилось отвергнуть нулевую гипотезу")
```

### Проверка гипотезы № 1 Уровень дохода влияет на отток

```python
# вызываю функцию для проверки гипотезы № 1
test_hyp(bank.query('churn==1 and est_salary.notna()')['est_salary'], \
         bank.query('churn==0 and est_salary.notna()')['est_salary'], 'est_salary', .05)
```

Исходя из полученного результата делаю вывод, что нулевая гипотеза № 1- "между клиентами, которые ушли и теми, которые остались нет значимой разницы в уровне дохода (столбец 'est_salary')", - подтвердилась. Это значит, что уровень дохода не влияет на отток. Данная проверка подтвердила результат корреляционного анализа.

- 'Segment' - сегмент пользователей;
- 'Total' - общее количество клиентов в сегменте;

### Составление сегментов

```python
# функция для определения комбинаций
def comb(sign, n):
    si = []
    for i in n: 
        si.extend(list(combinations(sign, i))) 
    return si     
```

```python
# создаю список признаков оттока
outflow_sign = ['credit_card == 1',
                'last_activity == 1', 
                'score_cat == "2_scor"', 
                'products == 2',
                'age_cat == "B"',
                'equity == 5']
n_element = [2, 3, 4] # количество элементов в комбинации

pd.options.display.max_colwidth = 130 # расширение вывода содержимого ячеек
sigm_comb = comb(outflow_sign, n_element) # получение списка кортежей комбининаций срезов
sigm_outflow = pd.DataFrame(sigm_comb) # создаю датафрейм со столбцами, в которых хранятся условия для срезов
sigm_outflow['Segment'] = pd.DataFrame({'Segment': sigm_comb}) # добавляю столбец с кортежем срезов

sigm_outflow.columns = ['p1', 'p2', 'p3', 'p4', 'Segment'] # переименовываю столбцы 
```

```python
sigm_outflow['Total'] = np.nan
sigm_outflow['Churn'] = np.nan

# создаю цикл для заполнения пустых столбцов
for index in range(len(sigm_outflow)):
    
    if sigm_outflow['p3'][index] == None:
        
        sigm_outflow.loc[index, 'Total'] = (bank
                                            .query(sigm_outflow['p1'][index])
                                            .query(sigm_outflow['p2'][index])['user_id'].count()
                                           )
        
        sigm_outflow.loc[index, 'Churn'] = (bank
                                            .query('churn == 1')
                                            .query(sigm_outflow['p1'][index])
                                            .query(sigm_outflow['p2'][index])['user_id'].count()
                                           )
        
    if sigm_outflow['p3'][index] != None and sigm_outflow['p4'][index] == None:
        
        sigm_outflow.loc[index, 'Total'] = (bank
                                            .query(sigm_outflow['p1'][index])
                                            .query(sigm_outflow['p2'][index])
                                            .query(sigm_outflow['p3'][index])['user_id'].count()
                                           )
        
        sigm_outflow.loc[index, 'Churn'] = (bank
                                            .query('churn == 1')
                                            .query(sigm_outflow['p1'][index])
                                            .query(sigm_outflow['p2'][index])
                                            .query(sigm_outflow['p3'][index])['user_id'].count()
                                           )
        
        
    if sigm_outflow['p4'][index] != None:
        
        sigm_outflow.loc[index, 'Total'] = (bank
                                            .query(sigm_outflow['p1'][index])
                                            .query(sigm_outflow['p2'][index])
                                            .query(sigm_outflow['p3'][index])
                                            .query(sigm_outflow['p4'][index])['user_id'].count())
        
        sigm_outflow.loc[index, 'Churn'] = (bank
                                            .query('churn == 1')
                                            .query(sigm_outflow['p1'][index])
                                            .query(sigm_outflow['p2'][index])
                                            .query(sigm_outflow['p3'][index])
                                            .query(sigm_outflow['p4'][index])['user_id'].count())
```

```python
# добавляю столбец 'Ratio' в котором будет храниться доля отточных клиентов

try:
    sigm_outflow['Ratio'] = sigm_outflow['Churn']/sigm_outflow['Total'] * 100
except ZeroDivisionError:
    sigm_outflow['Ratio'] = 0

# удаляю столбцы, по которым проходил цикл    
sigm_outflow = sigm_outflow.drop(columns=['p1', 'p2', 'p3', 'p4'])
sigm_outflow['Ratio'] = round(sigm_outflow['Ratio'], 2)
sigm_outflow[['Total', 'Churn']] = sigm_outflow[['Total', 'Churn']].astype('int64')

print('Количество сегментов:', len(sigm_outflow))
```
