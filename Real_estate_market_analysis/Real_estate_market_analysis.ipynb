## Предобработка данных

### Работа с названиями столбцов

```python
# привожу столбец cityCenters_nearest в "змеиный регистр"
data_estate = data_estate.rename(columns={'cityCenters_nearest' : 'city_centers_nearest'})
```

### Работа с пропусками

```python
# количество пропущенных значений для каждого столбца
data_estate.isna().sum()
```

В 14 столбцах есть пропуски. Они могли возникнуть из-за человеческого фактора, когда люди оформляли объявления. Также могли повлиять и технические сбои. Пропуски в следующих столбцах можно заполнить нулями:

- в balcony, скорее всего, пропуски, так как недвежимость без балкона.
- в parks_around3000 и ponds_around3000 пропуски, возможно, потому что в радиусе 3000 м нет парков/водоемов.

Пропуски в столбце is_apartment скорее всего означают, что эта недвижимость не апартаменты. Поэтому все пропуски можно заполнить значениями False.

Пропуски в столбце ceiling_height заполню медианой по всему датасету.

Пропуски в столбцах living_area и kitchen_area заполню произведением общей площади и коэффициента отношения площадей.

В ячейках столбца locality_name и floors_total есть пропуски и строки с такими значениями можно удалить, т.к. их мало и без этих значений они потеряли полезность для исследования. 

```python
# удаление строк с пропущенными значениями по столбцам locality_name и floors_total
data_estate = data_estate.dropna(subset=['locality_name'])
data_estate = data_estate.dropna(subset=['floors_total'])

# заполнение пропусков нулями в столбцах 'balcony', 'parks_around3000', 'ponds_around3000'
data_estate[['balcony', 'parks_around3000', 'ponds_around3000']] = \
data_estate[['balcony', 'parks_around3000', 'ponds_around3000']].fillna(value=0)

# заполнение пропусков в столбце is_apartment значением False
data_estate['is_apartment'] = data_estate['is_apartment'].fillna(value=False)

# заполнение пропусков в столбце ceiling_height
data_estate['ceiling_height'].fillna(data_estate['ceiling_height'].median(), inplace=True)
```

```python
# получаю коэффициенты отношения площадей
living_ratio = data_estate['living_area'].median() / data_estate['total_area'].median()
kitchen_ratio = data_estate['kitchen_area'].median() / data_estate['total_area'].median()

# заменяю пропуски в столбцах living_area и total_area 
data_estate['living_area'].fillna(living_ratio * data_estate['total_area'], inplace=True)
data_estate['kitchen_area'].fillna(kitchen_ratio * data_estate['total_area'], inplace=True)

# проверяю количество пропусков
data_estate.isna().sum()
```

Логичного решения для заполнения нулей в оставшихся столбцах не нашел. Оставляю пропуски.

### Работа с типами данных в датафрейме

При просмотре общей информации были выявлены столбцы, которые требуют изменения типа данных:

1. На тип int меняем last_price, так как значениями после запятой можно принебречь.
2. На тип int меняем: balcony, parks_around3000, ponds_around3000, т.к. в этих столбцах должны храниться целочисленные значения.
3. В тип datatime переводим first_day_exposition. В этом столбце находятся дата и время, поэтому тип object надо изменить.
4. С типа object на тип bool переводим is_apartment, так как в этом столбце находятся булевы значения.

Значения в столбцах 'floors_total' и 'days_exposition' тоже должны быть целочисленными, но в них содержатся пропуски и тип данных не изменить.

```python
# перевод типа столбцов 'last_price', 'floors_total', 'balcony', 'parks_around3000', 'ponds_around3000' в целочисленный
data_estate[['floor', 'floors_total', 'last_price', 'balcony', 'parks_around3000', 'ponds_around3000']] = \
data_estate[['floor', 'floors_total', 'last_price', 'balcony', 'parks_around3000', 'ponds_around3000']].astype('int64')
```

```python
# перевод в тип bool столбца is_apartment
data_estate['is_apartment'] = data_estate['is_apartment'].astype(bool)
```

```python
# перевод в тип datatime столбца first_day_exposition
data_estate['first_day_exposition'] = pd.to_datetime(data_estate['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

## Добавление в таблицу новых столбцов с расчётами

1. Добавление в датасет столбца price_sqm, в который войдет стоимость 1-го квадратного метра недвижимости.

```python
data_estate['price_sqm'] = data_estate['last_price'] / data_estate['total_area']

# привожу price_sqm в целочисленный тип
data_estate['price_sqm'] = data_estate['price_sqm'].astype(int)
```

2. Добавление столбца дня недели публикации объявления.

```python
data_estate['weekday'] = data_estate['first_day_exposition'].dt.weekday
```

3. Добавление столбца месяца публикации объявления.

```python
data_estate['month'] = data_estate['first_day_exposition'].dt.month
```

4. Добавление столбца года публикации объявления.

```python
data_estate['year'] = data_estate['first_day_exposition'].dt.year
```

5. Создание столбца с типом этажа квартиры: первый, последний, другой.

```python
# создаю функцию для добавления типа этажности недвижимости в новый столбец type_floor
def tp_floor(flr):
    if flr['floor'] == 1:
        return 'первый'
    elif flr['floor'] == flr['floors_total']:
        return 'последний'
    return 'другой'

data_estate['type_floor'] = data_estate.apply(tp_floor, axis=1)
```

6. Перевод расстояний от недвижимости до центра города из метра в километры.

```python
data_estate['city_centers_nearest_km'] = data_estate['city_centers_nearest'].where(data_estate['city_centers_nearest']=='NaN',\
                                                                    data_estate['city_centers_nearest']/1000).round()
data_estate.info()
```

**Вывод**

- Добавлен в датасет столбец price_sqm со стоимостью 1-го квадратного метра недвижимости.
- Добавлены столбцы: день, месяц и год публикации объявления.
- Добавлен столбец с типом этажности квартиры.
- Переведены расстояния от недвижимости до центра города из метра в километры.
