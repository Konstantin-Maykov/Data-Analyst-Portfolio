### Проверка значений на аномалии

#### Проверка на аномалии столбца critic_score

```python
games['critic_score'].unique()
```

Аномалий не обнаружено. Тип столбца можно перевести в тип Int64, т. к. нет ни одного дробного значения и есть значения "NaN", но лучше этого не делать, так как этот тип эксперементальный и могут появиться баги.

#### Проверка на аномалии столбца user_score

```python
games['user_score'].unique()
```

Выявлена аномалия - занчение 'tbd'. Из стороних источников узнал, что аббревиатура TBD  часто используется в сфере компьютерных игр и расшифровывается "To Be Determined", то есть "Будет определено". Это может означать, что оценка игры еще не установлена. Поэтому это значение можно приравнять к пропуску и заменить его значением "NaN".

```python
# замена 'tbd' на 'NaN'
games.loc[games['user_score'] == 'tbd', 'user_score'] = np.nan

# проверяю результат замены
games['user_score'].unique()
```

#### Проверка на аномалии столбца rating

```python
games['rating'].unique()
```

В этом столбце представлены возрастные рейтинги, присвоенные играм организацией ESRB. В столбце присутствует категория K-A - от детей до взрослых. Эта ктегория в 1998 году была переименована в категорию "Все" - "E". Поэтому значение K-A меняем на E.

```python
# замена K-A на E
games.loc[games['rating'] == 'K-A', 'rating'] = 'E'

# проверяю результат замены
games['rating'].unique()
```

### Продажи игр по платформам

```python
# создание сводной таблицы с суммарными продажами по каждой платформе
(games\
 .pivot_table(index='platform', values='total_sales', aggfunc='sum')\
 .sort_values(by='total_sales', ascending=False).reset_index()
)
```

Выбираю топ-10 платформ по сумме продаж за все года. В них входят: PS2, X360, PS3, Wii, DS, PS, PS4, GBA, PSP, 3DS. Строю графики с распределением продаж по годам по этим платформам.

```python
top_11_platform = ['PS2', 'X360', 'PS3', 'Wii', 'DS', 'PS', 'PS4', 'GBA', 'PSP', '3DS', 'PC']

for plat in top_11_platform:
    (games\
     .query('platform in @plat')\
     .pivot_table(index='year_of_release', values='total_sales', aggfunc='sum')\
     .plot(kind='bar', xlabel='Год', ylabel='Сумма продаж')
    )
    plt.title(plat, pad=30)
    plt.xticks(rotation=45);
```

При изучении графиков видно:
- большинство платформ появилось в начале 2000-х годов;
- в среднем платформы "живут" 10 лет.

### Определение актуального периода

Исходя из полученных результатов в предыдущем пункте можно сделать вывод, что в индустрии компьютерных игр тренды быстро меняются. Из этого следует, что для планирования рекламной компании на 2017 год не подойдет большой временной интервал, т. к. можно захватить уже изжившие себя платформы. Поэтому считаю, что лучше всего принять за актуальный период 3 года: с 2013 по 2016 год.

```python
# выполняю срез данных по периоду 20163-2016 гг
actual_period = games.query('year_of_release >= 2013')

# создаю сводную таблицу платформ за актуальный период
(actual_period\
 .pivot_table(index='platform', values='total_sales', aggfunc='sum')\
 .sort_values(by='total_sales', ascending=False)\
 .reset_index()
)
```

## Проверка гипотез

### Гипотезы по платформам Xbox One и PC

Нулевая гипотеза H0: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Альтернативная гипотеза H1: Средние пользовательские рейтинги платформ Xbox One и PC разные.

Задаем пороговое значение alpha 0.05.

В задаче у нас две генеральные совокупности: Xbox One и PC, поэтому для проведения теста использую метод .ttest_ind. Предполагаю, что дисперсии у совокупностей разные, поэтому equal_var будет принимать значение False.

```python
# удаление нулевых значений в столбце user_score
# actual_period = actual_period.dropna(subset='user_score')
```

```python
# расчет средних значений пользовательских рейтингов для XOne и PC
XOne = actual_period.query('(platform == "XOne") and (user_score.notna())')['user_score']
PC = actual_period.query('(platform == "PC") and (user_score.notna())')['user_score']
print('Средний пользовательский рейтинг платформы XOne', XOne.mean())
print('Средний пользовательский рейтинг платформы PC', PC.mean())
```

```python
# проверка гипотезы
alpha = .05
results = st.ttest_ind(XOne, PC, equal_var=False)
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Гипотеза подтвердилась. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

### Гипотезы по жанрам Action и Sports

Нулевая гипотеза H0: Средние пользовательские рейтинги жанров Action и Sports одинаковые.

Альтернативная гипотеза H1: Средние пользовательские рейтинги жанров Action и Sports разные.

Задаем пороговое значение alpha 0.05.

В задаче у нас две генеральные совокупности: жанры Action и Sports, поэтому для проведения теста используем метод .ttest_ind. Предполагаю, что дисперсии у совокупностей разные, поэтому equal_var будет принимать значение False.

```python
# расчет средних значений пользовательских рейтингов для Action и Sports
Action = actual_period.query('(genre == "Action") and (user_score.notna())')['user_score']
Sports = actual_period.query('(genre == "Sports") and (user_score.notna())')['user_score']
print('Средний пользовательский рейтинг платформы XOne', Action.mean())
print('Средний пользовательский рейтинг платформы PC', Sports.mean())
```

```python
# проверка гипотезы
alpha = .05
results = st.ttest_ind(Action, Sports, equal_var=False)
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Гипотеза не подтвердилась. Средние пользовательские рейтинги жанров Action и Sports разные.
